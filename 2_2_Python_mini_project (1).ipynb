```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Load the Titanic dataset from seaborn
titanic_df = sns.load_dataset('titanic')

# Display the first few rows of the dataset
print(titanic_df.head())
```

```python
plt.figure(figsize=(10, 6))
plt.hist(titanic_df['age'].dropna(), bins=30, edgecolor='k', alpha=0.7)
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df['age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
fig = px.histogram(titanic_df, x='age', nbins=30, title='Age Distribution of Titanic Passengers')
fig.update_layout(xaxis_title='Age', yaxis_title='Frequency')
fig.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(titanic_df['age'], titanic_df['fare'], alpha=0.7)
plt.title('Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='fare', data=titanic_df, alpha=0.7)
plt.title('Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
fig = px.scatter(titanic_df, x='age', y='fare', title='Age vs. Fare')
fig.update_layout(xaxis_title='Age', yaxis_title='Fare')
fig.show()
```

```python
plt.figure(figsize=(10, 6))
titanic_df.boxplot(column='fare', by='class', grid=False)
plt.title('Fare by Class')
plt.suptitle('')  # Suppress the automatic title
plt.xlabel('Class')
plt.ylabel('Fare')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='class', y='fare', data=titanic_df)
plt.title('Fare by Class')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.show()
```

```python
import plotly.express as px

# Create a box plot using Plotly Express
fig = px.box(titanic_df, x='class', y='fare', title='Fare by Class')
fig.update_layout(xaxis_title='Class', yaxis_title='Fare')
fig.show()
```
